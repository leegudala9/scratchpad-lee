```python
def get_perentage(tagged_entity):
    counts = Counter(elem[1] for elem in tagged_entity)
    number_of_occurences = counts["PERSON"]
    if number_of_occurences == 0:
        return 0
    percentage = number_of_occurences/len(tagged_entity)
    print("For this words {} the percentage was {}".format(tagged_entity, percentage*100))
    return percentage*100    
```

```python
for index in test_df.index:
    tokenized_word = word_tokenize(test_df.loc[index,"test"])
    print("\nDoing the tagging for word: ",test_df.loc[index,"test"])
    jammaku_jama = tagger.tag(tokenized_word)
    percentage = get_perentage(jammaku_jama)
    if percentage > 30:
        test_df.loc[index,"TEST_RESULTS"] = "ORGANIZATION"
    else:
        test_df.loc[index,"TEST_RESULTS"] = "NOT_AN_ORGANIZATION"

    
```
